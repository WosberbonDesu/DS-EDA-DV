##  Visualize Data

```python
df.hist(figsize = (20,20))
plt.show()
```

```python
category1 = ['acousticness','danceability', 'duration_ms', 'energy', 'explicit',
       'instrumentalness', 'key', 'liveness', 'loudness', 'mode',
       'popularity', 'speechiness', 'tempo', 'valence']
```

```python
for col in category1:
    y = df.groupby('year')[col].mean()
    x = y.index
    plt.figure(figsize=(15, 5))
    sns.set_style("whitegrid")
    sns.lineplot(x, y, data=df)
```

```python
plt.figure(figsize=(16, 8))
sns.set(style="whitegrid")
x = df.groupby("artists")["popularity"].mean().sort_values(ascending=False).head(20)
ax = sns.barplot(x.index, x)
ax.set_title('Top Artists with Popularity by Mean ')
ax.set_ylabel('Popularity')
ax.set_xlabel('Artists')
plt.xticks(rotation = 90)
```

## Comparison of Data and Ranking

* Data Analysis and and making ranking system this code show us popularity by artist

```python
plt.figure(figsize=(9, 3))
sns.set(style="whitegrid")
x = df.groupby("artists")["popularity"].sum().sort_values(ascending=False).head(20)
ax = sns.barplot(x.index, x)
ax.set_title('Top Artists with Popularity by Sum')
ax.set_ylabel('Popularity')
ax.set_xlabel('Artists')
plt.xticks(rotation = 90)
```

```python
arpop = pd.DataFrame(df.groupby('artists')['popularity'].sum()).sort_values('popularity').tail(10).reset_index()
fig = plt.figure(figsize=(15,5))
plt.bar( arpop['artists'],
         arpop['popularity'],
         width=0.45,
         color = ['#caf0f8','#ade8f4','#90e0ef','#48cae4','#00b4d8','#0096c7','#0077b6','#023e8a','#03045e','#14213d'])
plt.xticks(rotation=45,ha='right')
plt.title('Top 10 Most Popular Artists from 1921-2020',y=1.1,fontsize=20)
plt.xlabel('Artists')
plt.ylabel('Popularity (Ranges from 0 to 100)')
ax.axes.get_xaxis().set_visible(True)
```

```python
def plot_hist(variable):
    plt.figure(figsize=(9,3))
    plt.hist(df[variable],bins=50)
    plt.xlabel(variable)
    plt.ylabel("Frequency")
    plt.title("{} distrubition with hist".format(variable))
    plt.show()
```

### This code will show us how many music from which year


```python
numericVar = ["year"]
for i in numericVar:
    plot_hist(i)
```

### Songs Popularity(2000-2020)

```python
rel = df[df['year'] >= 2000]
variables = ['acousticness', 'danceability', 'energy','instrumentalness', 'key', 'liveness',
             'loudness', 'speechiness', 'tempo', 'valence']
year = range(2015,2021)

fig = plt.figure(figsize=(15,25))
for variable,num in zip(variables, range(1,len(variables)+1)):
    ax = fig.add_subplot(5,2,num)
    sns.scatterplot(variable, 'popularity', data=rel)
    plt.title('Relation between {} and Popularity'.format(variable))
    plt.xlabel(variable)
    plt.ylabel('Popularity')
fig.tight_layout(pad=0.5)
```

### How many songs by genre

```python

genre=df2[["genres","artists"]]
genre=genre[genre["genres"]!="[]"]
genre["genres"]=genre["genres"].str.replace("'", "")
genre["genres"]=genre["genres"].str.replace("[", "")
genre["genres"]=genre["genres"].str.replace("]", "")
genre_dict={}

genre_df_dict=genre["genres"].str.split(",")
for index, genre_list in genre_df_dict.iteritems():
    for genre_name in genre_list:
        if genre_name in genre_dict:
            genre_dict[genre_name]+=1
        else:
            genre_dict[genre_name]=1
            
genre_dict_sorted=sorted(genre_dict.items(), key= lambda x:x[1], reverse=True) 
genre_dict_sorted[1:20]




```
